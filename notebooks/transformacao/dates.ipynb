```r
atpD <- atp %>% 
    separate_wider_delim(Date, " - ", names=c("ds", "df"), too_few = "align_start", cols_remove=FALSE) %>% 
    mutate(tournament.date_start = ymd(ds), tournament.date_finish = ymd(df)) %>%
    mutate(tournament.duration = ifelse(
        tournament.date_finish %>% is.na(),
        0,
        tournament.date_finish - tournament.date_start
    ))
```

```r
options(scipen=999)
atpD %>% 
    group_by(tournament.duration) %>%
    summarise(n=n(), freq = n/nrow(atpD)*100) %>%
    arrange(-tournament.duration)
```

```r
# Numero de torneios diferentes por cada duração
```

```r
torneios <- atpD %>% group_by(Date, Tournament) %>% summarise(n.jogos=n(), .groups = 'drop', duration= first(tournament.duration)) %>% arrange(Tournament)
```

```r
options(scipen=999)
torneios %>% 
    group_by(duration) %>%
    summarise(n=n(), freq = n/nrow(torneios)*100) %>%
    arrange(-duration)
```
